```python
#custom dataset class,
class CustomIPDataset(torch.utils.data.Dataset):
    def __init__(self, dataFrame):
        self.instances = dataFrame['instance']
        self.target_values = dataFrame['nlps']

    def __len__(self):
        return len(self.instances)
    
    def __getitem__(self, idx):
        instance = self.instances[idx]
        target_value=self.target_values[idx]
        return instance, target_value
```

```python
#custom dataset

#dividing the data into test and train instances
#first, lets try by the naive first 4500 train, rest 500 is test
train_df = dataset.head(4500)
test_df= dataset.tail(500)
test_df.reset_index(drop=True, inplace=True)

training_data= CustomIPDataset(train_df)
test_data= CustomIPDataset(test_df)
```

```python
b_size=1

#dataloaders
train_dataloader = torch.utils.data.DataLoader(training_data, batch_size=b_size, shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_data, batch_size=b_size, shuffle=True)
```

```python
# Set the runtime environment to GPU in
if torch.cuda.is_available() is False:
    raise Exception("GPU device not found, runtime environment should be set to GPU")
print(f"Using GPU device: {torch.cuda.get_device_name(torch.cuda.current_device())}")

#GPU usage:
gpu=1

if gpu == -1:
    device = torch.device("cpu")
else:
    cuda = "cuda:"+str(gpu)
    device = torch.device(cuda if torch.cuda.is_available() else "cpu")
```

```python
#creating a more detailed attention and lstm based network

class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.hidden_size = hidden_size
        self.attention = nn.Linear(self.hidden_size, 1)
    
    def forward(self, inputs):
        attention_weights = self.attention(inputs)
        attention_weights = torch.softmax(attention_weights, dim=1)
        attended_inputs = attention_weights * inputs
        return attended_inputs

class AttentionModel(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(AttentionModel, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size)
        self.attention = Attention(hidden_size)
        self.linear1 = nn.Linear(hidden_size, 1)
        self.linear2 = nn.Linear(101,1)
        self.relu = nn.ReLU()
    
    def forward(self, inputs):
        lstm_out, _ = self.lstm(inputs)
        attended_out = self.relu(self.attention(lstm_out))
        before_pulling = self.relu(self.linear1(attended_out))
        output= self.relu(self.linear2(before_pulling.transpose(0,1)))
        output.view(1)
        return output
```

```python
#Initialize and move network to cuda
# define the model
model = AttentionModel(input_size=200, hidden_size=128).to(device)
print(model)
```

```python
# Initialize the loss function
#loss_fn = nn.CrossEntropyLoss()
#loss_fn = nn.CrossEntropyLoss()
#loss_fn = nn.CrossEntropyLoss()
loss_fn = nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-3)
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    train_loss= 0
    for batch, (x, y) in enumerate(dataloader):
        # Compute prediction and loss
        X = torch.Tensor(x).type(torch.FloatTensor).to(device)
        #print("X shape")
        #print(X.shape)
        X=X.view(101,200)
        #print(X.shape)
        Y = torch.Tensor(y).type(torch.FloatTensor).to(device)
        pred = model(X)
        #print("pred")
        #print(pred)
        Y = Y.view(b_size,1)
        #print("Y")
        #print(Y)
        loss = loss_fn(pred, Y)
        #print("loss")
        #print(loss)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        train_loss += loss_fn(pred, Y).item()
        
    train_loss /=num_batches
    print(f"Avg train loss: {train_loss:>8f}" )
    return train_loss

def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0

    with torch.no_grad():
        for x, y in dataloader:
            X = torch.Tensor(x).type(torch.FloatTensor).to(device)
            X= X.view(101,200)
            Y = torch.Tensor(y).type(torch.FloatTensor).to(device)
            Y = Y.view(b_size,1)
            pred = model(X)
            test_loss += loss_fn(pred, Y).item()

    test_loss /= num_batches
    print(f"Avg test loss: {test_loss:>8f}" )
    return test_loss
```

```python
epochs = 50
train_losses= []
test_losses=[]
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loss = train_loop(train_dataloader, model, loss_fn, optimizer)
    test_loss = test_loop(test_dataloader, model, loss_fn)
    
    train_losses.append(float(train_loss))
    test_losses.append(float(test_loss))
    print("Done!")
```

```python
def plot_errors(train_errors, test_errors):
  plt.plot(train_errors, label='Train Error')
  plt.plot(test_errors, label='Test Error')
  plt.xlabel('Epoch')
  plt.ylabel('Error')
  plt.legend()
  plt.show()

train_errors = train_losses
test_errors = test_losses
plot_errors(train_errors, test_errors)
```
